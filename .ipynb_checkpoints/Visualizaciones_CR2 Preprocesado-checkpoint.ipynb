### Tratamiento nulos y formateo de valores

```python
print("N° tuplas nulas precipitacion:",pr_data["codigo_estacion"].isnull().sum())
print("N° tuplas nulas caudal:",qflx_data["codigo_estacion"].isnull().sum())
print("N° tuplas nulas temperaturas:",tas_data["codigo_estacion"].isnull().sum())
print("N° tuplas nulas temperaturas minimas:",tasmin_data["codigo_estacion"].isnull().sum())
print("N° tuplas nulas temperaturas maximas:",tasmax_data["codigo_estacion"].isnull().sum())
```

```python
pr_data.iloc[:,0] =  pd.to_datetime(pr_data.iloc[:,0], format='%Y %m %d')
qflx_data.iloc[:,0] =  pd.to_datetime(qflx_data.iloc[:,0], format='%Y %m %d')
tas_data.iloc[:,0] =  pd.to_datetime(tas_data.iloc[:,0], format='%Y %m %d')
tasmin_data.iloc[:,0] =  pd.to_datetime(tasmin_data.iloc[:,0], format= "%Y %m %d")
tasmax_data.iloc[:,0] =  pd.to_datetime(tasmax_data.iloc[:,0], format= "%Y %m %d")

col1=pr_data.columns[0]
col2=qflx_data.columns[0]
col3=tas_data.columns[0]
col4=tasmin_data.columns[0]
col5=tasmax_data.columns[0]

pr_data = pr_data.set_index(pd.DatetimeIndex(pr_data[col1]))
qflx_data = qflx_data.set_index(pd.DatetimeIndex(qflx_data[col2]))
tas_data = tas_data.set_index(pd.DatetimeIndex(tas_data[col3]))
tasmin_data = tasmin_data.set_index(pd.DatetimeIndex(tasmin_data[col4]))
tasmax_data = tasmax_data.set_index(pd.DatetimeIndex(tasmax_data[col5]))

pr_data=pr_data.replace(['-9999',-9999], np.nan)
qflx_data=qflx_data.replace(['-9999',-9999], np.nan)
tas_data=tas_data.replace(['-9999',-9999], np.nan)
tasmin_data=tasmin_data.replace(['-9999',-9999], np.nan)
tasmax_data=tasmax_data.replace(['-9999',-9999], np.nan)

pr_data_M=pr_data.resample('M').mean()
qflx_data_M=qflx_data.resample('M').mean()
tas_data_M=tas_data.resample('M').mean()
tasmin_data_M=tasmin_data.resample('M').mean()
tasmax_data_M=tasmax_data.resample('M').mean()

pr_data_Y=pr_data.resample('Y').mean()
qflx_data_Y=qflx_data.resample('Y').mean()
tas_data_Y=tas_data.resample('Y').mean()
tasmin_data_Y=tasmin_data.resample('Y').mean()
tasmax_data_Y=tasmax_data.resample('Y').mean()
```

```python
#== Formateo "Codigo_estacion" y ordenamiento ==
data_stations['codigo_estacion'] = data_stations['codigo_estacion'].astype(str)
data_stations.sort_values(by=['codigo_estacion'], inplace=True)
data_stations = data_stations.reset_index(drop=True)
data_stations
```

```python
#== Revision de duplicados ==
data_stations.duplicated().value_counts()
```

```python
#== Borrado de duplicados
data_stations = data_stations.drop_duplicates().reset_index(drop=True)
data_stations.duplicated().value_counts()
```

```python
print("\033[1mDataset de estaciones:\033[0m")
print("   Cantidad de instancias:", data_stations.shape[0])
print("   Cantidad de atributos:", data_stations.shape[1])
data_stations
```

```python
#== Geolocalizacion ==
from geopy.geocoders import Nominatim
geoloc = Nominatim()
def geo(x,y,s):
  try:
    return geoloc.reverse(f'{x},{y}').raw["address"][s]
  except KeyError:
    return "otro"
  except TypeError:
    return "otro"
data_stations["pais"] = data_stations.apply(lambda x: geo(x[5],x[6],"country"), axis=1)
data_stations["region"] = data_stations.apply(lambda x: geo(x[5],x[6],"state"), axis=1)
```

### BigData

```python
#Para big data se usan las tablas transpuestas de mediciones

#== Reseteo a indice numerico (antes "codigo_estacion" era el indice) ==
pr_bigdata = pr_data_Mt.reset_index()
qflx_bigdata = qflx_data_Mt.reset_index()
tas_bigdata = tas_data_Mt.reset_index()
tasmin_bigdata = tasmin_data_Mt.reset_index()
tasmax_bigdata = tasmax_data_Mt.reset_index()

#== Correccion de formato (no me acuerdo exactamente xd) ==
pr_bigdata = pd.melt(pr_bigdata, id_vars="index", value_vars=pr_bigdata.columns[1:])
qflx_bigdata = pd.melt(qflx_bigdata, id_vars="index", value_vars=qflx_bigdata.columns[1:])
tas_bigdata = pd.melt(tas_bigdata, id_vars="index", value_vars=tas_bigdata.columns[1:])
tasmin_bigdata = pd.melt(tasmin_bigdata, id_vars="index", value_vars=tasmin_bigdata.columns[1:])
tasmax_bigdata = pd.melt(tasmax_bigdata, id_vars="index", value_vars=tasmax_bigdata.columns[1:])

#== Correccion de nombres (tampoco recuerdo bien) ==
pr_bigdata = pr_bigdata.rename(columns={'index':'codigo_estacion','codigo_estacion':'fecha'})
qflx_bigdata = qflx_bigdata.rename(columns={'index':'codigo_estacion','codigo_estacion':'fecha'})
tas_bigdata = tas_bigdata.rename(columns={'index':'codigo_estacion','codigo_estacion':'fecha'})
tasmin_bigdata = tasmin_bigdata.rename(columns={'index':'codigo_estacion','codigo_estacion':'fecha'})
tasmax_bigdata = tasmax_bigdata.rename(columns={'index':'codigo_estacion','codigo_estacion':'fecha'})

#== Join de cada tabla de mediciones transpuesta formateada con la tabla de datos de estaciones ==
pr_bigdata = pd.merge(pr_bigdata,data_stations,on='codigo_estacion')
qflx_bigdata = pd.merge(qflx_bigdata,data_stations,on='codigo_estacion')
tas_bigdata = pd.merge(tas_bigdata,data_stations,on='codigo_estacion')
tasmin_bigdata = pd.merge(tasmin_bigdata,data_stations,on='codigo_estacion')
tasmax_bigdata = pd.merge(tasmax_bigdata,data_stations,on='codigo_estacion')

#== Incorporacion de nueva columna para identificar el tipom de medicion ==
pr_bigdata.loc[:,"tipo_medicion"]="pr"
qflx_bigdata.loc[:,"tipo_medicion"]="qflx"
tas_bigdata.loc[:,"tipo_medicion"]="tas"
tasmin_bigdata.loc[:,"tipo_medicion"]="tasmin"
tasmax_bigdata.loc[:,"tipo_medicion"]="tasmax"

#== Union de todas en una gran tabla ==
bigdata = pd.concat([pr_bigdata, qflx_bigdata, tas_bigdata, tasmin_bigdata, tasmax_bigdata], axis=0)
bigdata
```

```python
#== Exploracion y deteccion de duplicados ==
print("\033[1mDataset de bigdata:\033[0m")
print("   Cantidad de instancias:", bigdata.shape[0])
print("   Cantidad de atributos:", bigdata.shape[1])
print("Duplicados observaciones bigdata:\n", bigdata.duplicated().value_counts())
```

```python
#== Descarga de big_data ==
bigdata.to_csv('big_data.csv',index=False)
files.download('big_data.csv')
```
